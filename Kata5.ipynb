Ejercicio 2

```python
# Almacenar las entradas del usuario
planetone = input("¿Cuál es la distancia respecto al sol del primer planeta?")
planettwo = input("¿Cuál es la distancia respecto al sol del segundo planeta?")
```

```python
# Convierte las cadenas de ambos planetas a números enteros
planetone = int(planetone)
planettwo = int(planettwo)
```

```python
# Realizar el cálculo y determinar el valor absoluto
# Convertir de KM a Millas
distance = (planettwo - planetone) * 0.621
print("Mi: " + str(abs(distance)))
```
