Solicita a un usuario que ingrese una lista de planetas. En un ejercicio posterior, agregar치s c칩digo que muestre la lista. Por ahora, crear치s solo el c칩digo que solicita al usuario la lista de planetas.

```python
# Declaramos 2 variabl
# es

list_planet = ''
planets = []
```

```python
# Escribe el ciclo while solicitado
# done para terminar de ingresar lista
while list_planet.lower() != 'done':
    if list_planet:
        planets.append(list_planet)
    list_planet = input('New planet ')
```

```python
# Escribe tu ciclo for para iterar en una lista de planetas

for planet in planets:
    print(planet)
```
