## Ejercicio 2
Calcularás tanto el número total de lunas en el sistema solar como el número promedio de lunas que tiene un planeta.

```python
# Planets and moons

planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
# Añade el código para determinar el número de lunas.

# se puede usar como una lista normal de Python. Puedes determinar el número de 
# elementos mediante len, e iterar a través de él mediante un ciclo for.
# El método keys() devuelve un objeto de lista que contiene todas las claves. 
planets = len(planet_moons.keys())
```

```python
# Agrega el código para contar el número de lunas. 
moons = planet_moons.values()
total_moons = 0
for moon in moons:
    total_moons = total_moons + moon

#Finalmente calcule el promedio dividiendo total_moons por planets e imprimiendo los resultados.
avarage = total_moons/planets
print(avarage)
```
