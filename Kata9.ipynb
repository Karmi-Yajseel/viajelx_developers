## Ejercicio 1
Construirás un informe de combustible que requiere información de varias ubicaciones de combustible en todo el cohete.

```python
# Función para leer 3 tanques de combustible y muestre el promedio
def tanks(first_tank, second_tank, third_tank):
    average_tank  = (first_tank + second_tank + third_tank) / 3
    ##Diseño de reporte
    return f""" Report:
    Total Avegare: {average_tank}%
    First tank: {first_tank}%
    Second tank: {second_tank}% 
    Third tank: {third_tank}% 
    """
```

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(tanks(80, 70, 85))
```

```python
# Mejora la legibilidad extrayendo el cálculo promedio de la función en una 
# nueva función para que el promedio se pueda hacer de forma independiente:*****

# Función promedio 
def average(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

average([80, 85, 81]) 
```

```python
# Actualiza la función
def tanks(first_tank, second_tank, third_tank):
    return f""" Report:
    Total Avegare: {average([first_tank, second_tank, third_tank])}%
    First tank: {first_tank}%
    Second tank: {second_tank}% 
    Third tank: {third_tank}% 
    """

## Test
print(tanks(88, 76, 70))
```

## Ejercicio 2
Construirás un informe de cohete que requiere varias piezas de información, como el tiempo hasta el destino, el combustible a la izquierda y el nombre del destino. 

```python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno
# La función es problemática porque no se puede adaptar para minutos adicionales o tanques adicionales de combustible.
# Hazlo flexible permitiendo cualquier número de pasos basados en el tiempo y cualquier número de tanques.

def report_mission(pre_launch_time, flight_time, destination, external_tank, main_tank):
    return f"""
    Mission to {destination}
    Total travel time: {pre_launch_time + flight_time} minutes
    Total fuel left: {external_tank + main_tank} gallons
    """
print(report_mission(14, 51, "Moon", 200000, 300000))
```

```python
#La sintaxis para usar argumentos de variable es agregar un asterisco único como prefijo (*) antes del nombre del argumento.
#Para que una función acepte cualquier número de argumentos de palabra clave, debe usar una sintaxis similar. En este caso, se requiere un asterisco doble.
# --- En lugar de usar *args y **kwargs, aprovecha el hecho de que puedes usar cualquier nombre que desees.

# Escribe tu nueva función de reporte considerando lo anterior
def mission_report(destination, *minutes, **fuel_reservoirs):
    return f"""
    Mission to {destination}
    Total travel time: {sum(minutes)} minutes
    Total fuel left: {sum(fuel_reservoirs.values())}
    """

print(mission_report("Moon", 10, 15, 51, main=300000, external=200000))
```

```python
# El combustible que queda en los tanques es específico de cada tanque, actualiza la función.
def mission_report(destination, *minutes, **fuel_reservoirs):
    main_report = f"""
    Mission to {destination}
    Total travel time: {sum(minutes)} minutes
    Total fuel left: {sum(fuel_reservoirs.values())}
    """

    # para usar el nombre de cada tanque en el informe.
    for tank_name, gallons in fuel_reservoirs.items():
        main_report += f"{tank_name} tank --> {gallons} gallons left\n"
    return main_report

print(mission_report("Moon", 8, 11, 55, main=300000, external=200000))
```
